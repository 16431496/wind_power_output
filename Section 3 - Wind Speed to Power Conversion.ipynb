### From the power curves, wind speeds are record every 0.5 m/s and the power is only given for every increase in wind speed by 0.5 m/s. The kernal below changes that to every 0.1 m/s and gives more options for power output.

```python
if num ==0: 
     print('Error. Rerun code.')
else:
    x = dfpo.index.to_numpy()
    y = dfpo[turbine].to_numpy()
    x_new = np.linspace(0,26,260001)
    y_new = np.interp(x_new,x,y)
    power_dict = dict(zip(x_new.astype(float).round(4), y_new.astype(float).round(4))) #creates power dictionary
```

```python
#kernal gets windspeed at hub height, power at every windspeed and the total power from all wind turbines
if num ==0: 
     print('Error. Rerun code.')
else:
    powerlist = []

    dfws['windspeed at a hub height of %s m (m/s)' %str(z4)] = (dfws.windspeed * ((ln(z4/z0))/(ln(z1/z0)))).round(4)

    for i in range(0, len(dfws)):
        

        if (dfws['windspeed at a hub height of %s m (m/s)' %(str(z4))].iloc[i]) <= 25.0:
            powerlist.append(power_dict[dfws['windspeed at a hub height of %s m (m/s)' %(str(z4))].iloc[i]])
        else:
            powerlist.append(0)
            
    dfws['%s power at a hub height of %s m (kW)' %(turbine,str(z4))] = powerlist

    for i in range(0,len(powerlist)):
        powerlist[i] = (powerlist[i]*num)/1000.0

    dfws['%s turbines for %s power at a hub height of %s m (MW)' %(num,turbine,str(z4))] = powerlist

    dfws = dfws.drop(['windspeed at a hub height of %s m (m/s)' %(str(z4))], axis=1)
```
